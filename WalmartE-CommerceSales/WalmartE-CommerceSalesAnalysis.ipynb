# Walmart E-Commerce Sales Analysis
This notebook contains a very basic analysis of a dataset containing Walmart e-commerce sales data. The dataset was downloaded from kaggle and can be found [here](https://www.kaggle.com/datasets/devarajv88/walmart-sales-dataset).

```python
import pandas as pd

df = pd.read_csv('walmartsales.csv')
df.head()
```

```python
print(f"The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# statistical info
df.describe()
```

```python
# datatype info
df.info()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# check for duplicates
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt

# Group by 'Age' and sum 'Purchase' column
age_group_sales = df.groupby('Age')['Purchase'].sum()

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Create a pie chart
age_group_sales.plot(kind='pie', autopct='%1.1f%%', ax=ax)

# Add a title
ax.set_title('Sales Distribution Among Different Age Groups')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by 'Age' and sum 'Purchase' column
age_group_customers = df.groupby('Age').size()

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Create a pie chart
age_group_customers.plot(kind='pie', autopct='%1.1f%%', ax=ax)

# Add a title
ax.set_title('Customer Distribution Among Different Age Groups')

# Show the plot
plt.show()
```

```python
# Group by 'Age' and sum 'Purchase' column
age_group_sales = df.groupby('Age')['Purchase'].sum()

# Group by 'Age' and sum 'Purchase' column
age_group_customers = df.groupby('Age').size()

# Calculate average sales per customer
average_sales_per_customer = age_group_sales / age_group_customers

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar chart
average_sales_per_customer.plot(kind='bar', ax=ax, color='skyblue')

# Add a title and labels
ax.set_title('Average Sales per Customer for Each Age Group')
ax.set_xlabel('Age Group')
ax.set_ylabel('Average Sales per Customer')

# Show the plot
plt.show()
```

```python
# Group by 'City_Category' and sum 'Purchase' column to get total sales
total_sales_city = df.groupby('City_Category')['Purchase'].sum()

# Group by 'City_Category' and count the rows to get total customers
total_customers_city = df.groupby('City_Category').size()

# Calculate average sales per customer
average_sales_per_customer_city = total_sales_city / total_customers_city

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar chart
average_sales_per_customer_city.plot(kind='bar', ax=ax, color='skyblue')

# Add a title and labels
ax.set_title('Average Sales per Customer for Each City Category')
ax.set_xlabel('City Category')
ax.set_ylabel('Average Sales per Customer')

# Show the plot
plt.show()
```

```python
# Group by 'City_Category' and sum 'Purchase' column to get total sales
customers_gender = df.groupby('Gender').size()

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Add a title
ax.set_title('Customer Distribution based on Gender')

# Create a pie chart
customers_gender.plot(kind='pie', autopct='%1.1f%%', ax=ax);
```

```python
# Group by 'Stay_In_Current_City_Years' and sum 'Purchase' column to get total sales
total_sales_stay = df.groupby('Stay_In_Current_City_Years')['Purchase'].sum()

# Group by 'Stay_In_Current_City_Years' and count the rows to get total customers
total_customers_stay = df.groupby('Stay_In_Current_City_Years').size()

# Calculate average sales per customer
average_sales_per_customer_stay = total_sales_stay / total_customers_stay

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar chart
average_sales_per_customer_stay.plot(kind='bar', ax=ax, color='skyblue')

# Add a title and labels
ax.set_title('Average Sales per Customer for Each Stay in Current City')
ax.set_xlabel('Stay in Current City (Years)')
ax.set_ylabel('Average Sales per Customer')

# Show the plot
plt.show()
```

```python
# Check the biggest purchase, the biggest spenders and the most popular product
biggest_purchase = df['Purchase'].max()
biggest_spenders = df.groupby('User_ID')['Purchase'].sum().max()
most_popular_product = df['Product_ID'].mode()[0]

print(f"The biggest purchase was ${biggest_purchase}.")
print(f"The biggest spender spent ${biggest_spenders}.")
print(f"The most popular product was {most_popular_product}.")

# Calculate the average sale and the average sale per customer
average_sale = df['Purchase'].mean()
average_sale_per_customer = df.groupby('User_ID')['Purchase'].sum().mean()

print(f"The average sale was ${average_sale:.2f}.")
print(f"The average sale per customer was ${average_sale_per_customer:.2f}.")
```
